```python
#loading required libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#loading the ipl 2017 data set
ipl = pd.read_csv('matches.csv')
```

```python
#having a glance at the first five records of the dataset
ipl.head()
```

```python
#looking for number of rows and columns
ipl.shape 
```

```python
#getting the frequency of most man of the match awards
ipl['player_of_match'].value_counts()
```

```python
#getting the top 10 players with man of the match awards
ipl['player_of_match'].value_counts()[0:10]
```

```python
#getting the top 5 players with man of the match awards
ipl['player_of_match'].value_counts()[0:5]
```

```python
#getting the top 5 players with man of the match awards with only names
list(ipl['player_of_match'].value_counts()[0:5].keys())
```

```python
#getting bar-plot of the top 5 players with man of the match awards
plt.figure(figsize=(8,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]),color="b")
plt.show()
```

```python
#getting the frequency of result column 
ipl['result'].value_counts()
```

```python
#finding out the number of toss win w.r.t each team
ipl['toss_winner'].value_counts()
```

```python
#extracting the results where team won by batting first
batting_first = ipl[ipl['win_by_runs']!=0]
```

```python
#looking at the head
batting_first.head()
```

```python
#making histogram
plt.figure(figsize=(5,5))
plt.hist(batting_first['win_by_runs'])
plt.title("Distribution of runs")
plt.xlabel('Runs')
plt.ylabel('Matches')
plt.show()
```

```python
#finding out the number of wins win w.r.t each team after batting first
batting_first['winner'].value_counts()
```

```python
#making bar-plot for top 3 teams with the most wins after batting first
plt.figure(figsize=(6,6))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color=["blue","yellow","red"])
plt.show()
```

```python
#making a pie chart
plt.figure(figsize=(20,20))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.legend(title="Ipl Teams")
plt.show()
```

```python
#extracting the results where team won by batting second
batting_second = ipl[ipl['win_by_wickets']!=0]
```

```python
#looking at the head
batting_second.head()
```

```python
#making histogram
plt.figure(figsize=(5,5))
plt.hist(batting_second['win_by_wickets'],bins=30)
plt.xlabel('Wickets')
plt.ylabel('Matches')
plt.show()
```

```python
#finding out frequency of number of wins w.r.t each time after batting second
batting_second['winner'].value_counts()
```

```python
#making a bar-plot for top 3 teams with most wins after batting second
plt.figure(figsize=(6,6))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),list(batting_second['winner'].value_counts()[0:3]),color=["purple","blue","red"])
plt.show()
```

```python
#making a pie chart for distribution of most wins after batting second 
plt.figure(figsize=(20,20))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.legend(title="Ipl Teams")
plt.show()
```
